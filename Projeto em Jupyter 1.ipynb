# 4. Análise Univariada

```python
_= plt.hist(rec_humanos.Age)
_= plt.title("Histograma de Idade")
# Voltamos até a parte 2, carregando bibliotecas, para modificar o layout do gráfico utilizando " plt.style.use('ggplot') "
```

```python
_= plt.hist(rec_humanos.MonthlyIncome)
_= plt.title("Histograma de Salário")
```

```python
rec_humanos.Gender.value_counts(0)
# Comando para ver a distribuicão de genero
# Se modificarmos o campo () para (1), ele entende como porcentagem...
```

```python
rec_humanos.Gender.value_counts(1)
# Ao modificarmos o campo () para (1), ele entende como porcentagem...
```

 Por ser uma empresa que contrata muitas pessoas da área de STEM (science, technology, engineering, mathematics), é esperado que tenha mais homens do que mulheres.

```python
rec_humanos.EducationField.value_counts(0)
# Comando para ver a distribuicão do Campo Educacional das pessoas
```

```python
rec_humanos.EducationField.value_counts(1)
```

```python
_= rec_humanos.EducationField.value_counts(1).plot.bar()
_= plt.title("Headcount por Campo de Educacão")
# Lembrando que o ( _= ) antes de criar o gráfico evita que aparece o nome ( <AxesSubplot:> ).
```

Nota-se que a maioria dos funcionários é formada na área de Life Sciences.  
Em seguida, temos muitos funcionários da área Medical.  
A área com menos funcionários formados é a de Human Resources.

Espera-se encontrar salários maiores para homens, porém, no caso da IBM, parece que não há Gender Gap.
